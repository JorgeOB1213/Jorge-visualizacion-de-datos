```python
df=pd.read_csv('datasets/Cartwheeldata.csv')
df.head(2)
```

```python
import seaborn as sns

sns.lmplot(
    x='Wingspan',
    y='CWDistance',
    data = df,
    fit_reg = False,
    hue = 'Gender'
    
)
plt.show()
```

```python
sns.swarmplot(
    x = 'Gender',
    y = 'CWDistance',
    data = df
)
plt.show()
```

```python
sns.boxplot(
    data = df.loc[:,['Age','Height','Wingspan','CWDistance','Score']]
)
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)# 1 renglos, 2 columns, position 1 
sns.boxplot(
    data = df.loc[df['Gender']=='F','Score']
)

ax2 = fig.add_subplot(122)
sns.boxplot(
    data2 = df.loc[df['Gender']=='M', 'Score']
)
plt.show()
```
