Como indentificar generacion de consolas?

```python
def create_category(gen):
    if (gen>=1980) &(gen<=1988):
        return "Tercera generación" 
    if (gen>=1989) &(gen<=1993):
        return "Cuarta generación"
    if (gen>=1994) &(gen<=2004):
        return "Quinta generación"
    if (gen>=2005) &(gen<=2012):
        return "Sexta generación"
    if (gen>=2013) &(gen<=2020):
        return "Septima generación"
    
#Aqui configuramos rangos para alinearlos con la fecha que salio cada videojuego para determinar de que generacion de videojuego
#es
    
    
```

```python
df["gen"]=df["Year_of_Release"].apply(lambda x:create_category(x))
df
#en base a los parametros pasados ya podremos ver de que generaicon es cada videojuegos (generacion de consla
```
